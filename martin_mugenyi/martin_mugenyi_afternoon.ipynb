Qn 1. Write a Pandas program to compare the elements of two Pandas series. 
Sample Series [4,65,436,3,9], [7,0,3,897,9]


```python
series_1 = pd.Series([4, 65, 436, 3, 9])
series_2 = pd.Series([7, 0, 3, 897, 9])

# comparing two series
compared_series = series_2 == series_1

print(compared_series)
```

Qn 2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
Sample Series: [2,4,6,8,14], [1,3,5,7,9]


```python
series_3 = pd.Series([2, 4, 6, 8, 14])
series_4 = pd.Series([1,3,5,7, 9])

# Adding
add = series_3 + series_4

# Subtracting
subtract = series_3 - series_4

# Division
divide = series_3 / series_4

# Multiplying
multiply = series_3 * series_4

results = pd.DataFrame(
    {
        'Series_3': series_3,
        'Series_4': series_4,
        'Addition': add,
        'Subtraction': subtract,
        'Multiplication': multiply,
        'Division': divide,
    }
)

results
```

qn 5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)

```python
df = pd.read_csv('AfricaCupofNationsMatches.csv', encoding="latin 1")

HomeTeamz = df['HomeTeamGoals']

Home_Teams = pd.Series(HomeTeamz)
# most frequent
most_freq = Home_Teams.mode()[0]
print("Modal value: ", most_freq)

# replacing modal mark with "Other"
Home_Teams[Home_Teams == most_freq] = 'Other'

# Display data
Home_Teams
```
